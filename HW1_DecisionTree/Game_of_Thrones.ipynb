```python
import pandas as pd
import numpy as np
import scipy
import pydot
from math import isnan
from sklearn import tree
from sklearn.externals.six import StringIO
```

## 1)將資料讀取進來(可用pandas套件)

```python
death_data = pd.read_csv("character-deaths.csv")
death_data
```

## 2)將資料處理(請用程式碼處理)

### 2-1把空值以0替代

```python
len(death_data.columns), len(death_data), death_data.columns
```

```python
for i in xrange(2, len(death_data.columns)):
    for j in xrange(0, len(death_data)):
        if isnan(death_data[death_data.columns[i]][j]):
            death_data[death_data.columns[i]][j] = 0
```

```python
death_data
```

### 2-2Death Year , Book of Death , Death Chapter三者取一個，將有數值的轉成1

```python
# delete Death Year and Book of Death
death_data = death_data.drop(death_data.columns[[2,3]], axis=1)
```

```python
len(death_data.columns), len(death_data), death_data.columns
```

```python
for j in xrange(0, len(death_data)):
    if death_data[death_data.columns[2]][j] != 0:
        death_data[death_data.columns[2]][j] = 1
```

```python
death_data
```

### 2-3將Allegiances底下的家族轉成dummy的特徵(底下有幾種分類就會變成幾個特徵，值是0或1，本來的資料集就會再增加好幾十種特徵)

```python
# get Allegiances data
dummy = []
for i in xrange (0, len(death_data)):
    dummy.append(death_data[death_data.columns[1]][i])
```

```python
# transfer Allegiances data to dummy feature
tmp_label = []
for x in sorted(dummy):
    tmp_label.append(x)
dummy_label = [tmp_label[0]]

for i in xrange(1, len(tmp_label)):
    if tmp_label[i-1] != tmp_label[i]:
        dummy_label.append(tmp_label[i])
```

```python
len(dummy_label), dummy_label
```

```python
# transfer dummy feature to 0 or 1
for i in xrange (0, len(dummy_label)):
    death_data[str(dummy_label[i])] = pd.Series(np.zeros(len(death_data)), index=death_data.index)

for i in xrange(0, len(dummy_label)):
    for j in xrange(0, len(death_data)):
        if death_data[death_data.columns[1]][j] == dummy_label[i]:
            death_data[death_data.columns[i+11]][j] = 1
```

```python
death_data
```

## 3)使用python 中scikit-learn上的decision tree(前75%是訓練集、後25%當測試集,可以先試著將網頁範例(iris)跑出來在使用這次作業的資料集)

```python
# transform dataframe to np array
death_matrix = death_data.as_matrix()

# delete name and Allegiances, transform the data type to int
death_matrix = scipy.delete(death_matrix, [0,1], 1)
death_matrix = death_matrix.astype(int)

train_len = int(len(death_data)/4*3)
test_len = len(death_data) - train_len
# data set
train = death_matrix[0:train_len]
test = death_matrix[train_len::]
train_data = train[0:,1:train.shape[1]]
train_label = train[:, 0]
test_data = test[0:, 1:test.shape[1]]
test_label = test[:, 0]
```

```python
# train decision tree model
clf = tree.DecisionTreeClassifier()
clf = clf.fit(train_data, train_label)
```

```python
# testing
pred = clf.predict(test_data)
pred
```

## 4)計算Precision Rate , Recall Rate , Accuracy

```python
ConfusionMatrix = np.zeros(4, dtype=float)
# (TP, FP, FN, TN)
for i in xrange (0, len(pred)):
    if test_label[i]==1 and pred[i]==1:
        ConfusionMatrix[0] = ConfusionMatrix[0] + 1
    elif test_label[i]==0 and pred[i]==1:
        ConfusionMatrix[1] = ConfusionMatrix[1] + 1
    elif test_label[i]==1 and pred[i]==0:
        ConfusionMatrix[2] = ConfusionMatrix[2] + 1
    else:
        ConfusionMatrix[3] = ConfusionMatrix[3] + 1

PrecisionRate = ConfusionMatrix[0] / (ConfusionMatrix[0] + ConfusionMatrix[1])
RecallRate = ConfusionMatrix[0] / (ConfusionMatrix[0] + ConfusionMatrix[2])
Accuracy = (ConfusionMatrix[0] + ConfusionMatrix[3]) / len(pred)

print 'Precision Rate:', PrecisionRate
print 'Recall Rate:', RecallRate
print 'Accuracy:', Accuracy
```

## 5)產出決策樹的圖
  p.s 記得限制樹的深度，以免結果無法顯示

```python
dot_data = StringIO()
tree.export_graphviz(clf, out_file=dot_data) 
graph = pydot.graph_from_dot_data(dot_data.getvalue())
graph.write_pdf("death.pdf")
graph.write_jpeg("death.jpeg")
```
