```python
import jieba
import _uniout
import numpy as np

from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
corpus = []
with open('content.txt', 'r') as f:
    for line in f:
        corpus.append(" ".join(jieba.cut(line.split(',')[0], cut_all=False)))
```

```python
print _uniout.unescape(str(corpus), 'utf8')
```

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(corpus)
print tfidf.shape
```

```python
words = vectorizer.get_feature_names()
for i in xrange(len(corpus)):
    print '----Document %d----' % (i)
    for j in xrange(len(words)):
        if tfidf[i,j] > 1e-5:
              print words[j].encode('utf-8'), tfidf[i,j]
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(tfidf)
```

```python
ans = np.array(100,dtype=int)
ans = kmeans.predict(tfidf)
```

```python
print '------第1類------'
for i in xrange(0, 100):
    if ans[i]==0:
        print corpus[i]
```

```python
print '------第2類------'
for i in xrange(0, 100):
    if ans[i]==1:
        print corpus[i]
```

```python
print '------第3類------'
for i in xrange(0, 100):
    if ans[i]==2:
        print corpus[i]
```

```python
score = -kmeans.score(tfidf)
print 'n_clusters=3, score=', score
```

```python
for i in xrange(4, 11):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(tfidf)
    score = -kmeans.score(tfidf)
    print 'n_clusters=', i, ', score=', score
```
